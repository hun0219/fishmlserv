```python
from sklearn.neighbors import KNeighborsClassifier
kn = KNeighborsClassifier()
```

```python
fish_data = [
    [25.4,242.0],     # data(),  
    [26.3,290.0],
    ....
    (34)
    [9.8,6.7],   #  data(),  
    [10.5,7.5]
]
fish_target = [ ]
```

```python
#   list 
fish_data = []             #fish_data  list

for l, w in zip(length, weight):     #zip :    .  length, weight
    fish_data.append([l, w])    #append -     [25.4, 242.0]

print(fish_data)
```

```python
fish_target = [1] * 35 + [0] *14   #    35*14    # 1 0 
                                #[25.4, 242.0] == [1] # (35)
                                #[12.0, 9.8] == [0]  # (14)

print(fish_target)
```

```python
#  ML     fiting.  kn  .     # fish_data, fish_target   [25.4, 242.0] [1] == 
kn.fit(fish_data, fish_target)
```

```python
# predict
r = kn.predict([[30.1, 600.123]])     # 30.1,  600.123    
print(r)   # 
```
